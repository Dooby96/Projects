<p style="text-align: center;"><span style="font-family: Helvetica; font-size: 24px;"><span style="color: rgb(0, 156, 222);"><strong>Part I:</strong></span> Exploratory analysis</span></p>

```python
import pandas as pd
pd.options.display.float_format = '{:,.2f}%'.format
```

<p style="text-align: left;"><span style="font-family: Helvetica; font-size: 20px;"><span style="color: rgb(0, 156, 222);"><strong>1.1</strong></span></span><span style="font-family: Helvetica; font-size: 20px;">&nbsp;Import and Index Selection</span></p>

```python
df = pd.read_csv('train.csv', index_col='PassengerId')
df_original = df.copy()

df.head()
```

<p style="text-align: left;"><span style="font-family: Helvetica; font-size: 20px;"><span style="color: rgb(0, 156, 222);"><strong>1.2</strong></span></span><span style="font-family: Helvetica; font-size: 20px;">&nbsp;Data Cleaning</span></p>

<p style="text-align: left;"><span style="font-family: Helvetica; font-size: 15px;"><span style="color: rgb(0, 0, 0);"><strong></strong></span><span style="color: rgb(0, 0, 0);">&nbsp;Reformat some columns</span></p>

```python
df['Survived'] = df['Survived'].astype(bool)
df['SibSp'] = df['SibSp'].astype(bool)
df['Sex'] = df['Sex'].str.capitalize()

df.head()
```

<p style="text-align: left;"><span style="font-family: Helvetica; font-size: 20px;"><span style="color: rgb(0, 156, 222);"><strong>2.1</strong></span></span><span style="font-family: Helvetica; font-size: 20px;">&nbsp;Percentage of Survivors out of all</span></p>

<p><span style="font-family: Helvetica; font-size: 15px;">Using Panda's groupby() function we can easily make a table containing the sum of all people that survived, grouping them by sex. Then we obtain a series which values can be divided by the total number of passengers to obtain a percentage (%). Finnally we multiply results by 100 to remove decimal format</span></p>
<p><span style="font-family: Helvetica; font-size: 15px;"></span></p>

```python
survived = df.groupby("Survived")["Name"].count()/df["Name"].count()*100
survived
```

<p style="text-align: left;"><span style="font-family: Helvetica; font-size: 20px;"><span style="color: rgb(0, 156, 222);"><strong>2.2</strong></span></span><span style="font-family: Helvetica; font-size: 20px;">&nbsp;Percentage of men and women out of total</span></p>

<p><span style="font-family: Helvetica; font-size: 15px;">Similarly as in previous solution, we use groupby() to sum passengers in the two different sex groups. Then we divide the sums by the total number of passengers and multiply by 100</span></p>
<p><span style="font-family: Helvetica; font-size: 15px;"></span></p>

```python
sex = df.groupby("Sex")["Name"].count()/df["Name"].count()*100
sex
```

<p style="text-align: left;"><span style="font-family: Helvetica; font-size: 20px;"><span style="color: rgb(0, 156, 222);"><strong>3.1</strong></span></span><span style="font-family: Helvetica; font-size: 20px;">&nbsp;Percentage of survivors by sex</span></p>

<p><span style="font-family: Helvetica; font-size: 15px;">The same method is repeated for this question. In this case the difference is that we need to divide each group by its corresponding total count, so we also apply groupby("Sex") function in the denominator of the formula</span></p>

```python
sexSurvived = df.groupby("Sex")["Survived"].sum()/df.groupby("Sex")["Name"].count()*100
sexSurvived
```

<p style="text-align: left;"><span style="font-family: Helvetica; font-size: 20px;"><span style="color: rgb(0, 156, 222);"><strong>3.2</strong></span></span><span style="font-family: Helvetica; font-size: 20px;">&nbsp;Sex distribution of survivors</span></p>

<p><span style="font-family: Helvetica; font-size: 15px;">Using groupby() we obtain the sum of survivors grouped by sex. Then we divide each group by the sum of survivors of the corresponding sex</span></p>

```python
survivedSex = df.groupby("Sex")["Survived"].sum()/df[df["Survived"]==True]["Name"].count()*100
survivedSex
```

<p style="text-align: left;"><span style="font-family: Helvetica; font-size: 20px;"><span style="color: rgb(0, 156, 222);"><strong>4.</strong></span></span><span style="font-family: Helvetica; font-size: 20px;">&nbsp;Probability Dataframe by Sex and Survival - Total = 1</span></p>

<p><span style="font-family: Helvetica; font-size: 15px;">In order to do do this easily with Pandas, we can insert a new column with the opposite result of "Survived". This column is called "Not Survived". As "Survived" is a boolean, we can inverse it by just substacting it to 1.</span></p>

```python
df.insert(1,"Not Survived",1-df["Survived"])
df['Not Survived'] = df['Not Survived'].astype(bool)
df.head(1)
```

<p><span style="font-family: Helvetica; font-size: 15px;">The required Dataframe can be obtained with the same methods used previously. The only difference in this case is that we will summarize two columns instead of one. These columns will be "Survived" and "Not Survived".</span></p>

```python
df_1 = df.groupby(["Sex"])[["Survived", "Not Survived"]].sum()/df["Name"].count()*100
df_1
```

<p style="text-align: left;"><span style="font-family: Helvetica; font-size: 20px;"><span style="color: rgb(0, 156, 222);"><strong>5.</strong></span></span><span style="font-family: Helvetica; font-size: 20px;">&nbsp;Probability Dataframe by Sex, Class and Survival - Total = 1</span></p>

<p><span style="font-family: Helvetica; font-size: 15px;">It is possible to add an additional grouping level by just including the the column with the desired categories in the index for the groupby() function</span></p>

```python
df_2 = df.groupby(["Sex", "Pclass"])[["Survived", "Not Survived"]].sum()/df["Name"].count()*100
df_2
```

<p style="text-align: center;"><span style="font-family: Helvetica; font-size: 24px;"><span style="color: rgb(0, 156, 222);"><strong>Part II:</strong></span> Two Strategies for Missing Value Imputation</span></p>

<p style="text-align: left;"><span style="font-family: Helvetica; font-size: 20px;"><span style="color: rgb(0, 156, 222);"><strong>1.1.</strong></span></span><span style="font-family: Helvetica; font-size: 20px;">&nbsp;Impute the median age for each "Title" group</span></p>

<p><span style="font-family: Helvetica; font-size: 15px;">First a new column "Title" is created by extracting the title string from the "Name" column</span></p>

```python
df.insert(3,"Title",df["Name"].str.split(",").str.get(1).str.split(".").str.get(0))
df.head()
```

<p><span style="font-family: Helvetica; font-size: 15px;">A table is created grouping "Age" by average for each corresponding "Title" group</span></p>

```python
ageByTitle=df.groupby("Title", as_index=False)["Age"].median()
ageByTitle=ageByTitle.rename(columns={"Age":"Age_MedianStr1"})
ageByTitle
```

<p><span style="font-family: Helvetica; font-size: 15px;">A new dataframe "fullAge" is defined to complete it with new columns. It is built by joining the original dataframe with "ageByTitle". As result of the merge "Age_MedianStr1" is now a column of the dataframe, so for each passenger it is possible to see the median age of his/her "Title" group. </span></p>
<p><span style="font-family: Helvetica; font-size: 15px;"></span>Then "Age_Str1" is created. This column contains all age values. When the "Age" value is null, it is filled by the median of "Title" value found in "Age_MedianStr1". This is achieved by using .combine_first() method</p>

```python
fullAge=pd.merge(df,ageByTitle,how='left',on='Title')
fullAge["Age_Str1"]=fullAge["Age"].combine_first(fullAge["Age_MedianStr1"])
```

<p style="text-align: left;"><span style="font-family: Helvetica; font-size: 20px;"><span style="color: rgb(0, 156, 222);"><strong>1.2.</strong></span></span><span style="font-family: Helvetica; font-size: 20px;">&nbsp;Impute average age for each "Title" group</span></p>

<p><span style="font-family: Helvetica; font-size: 15px;">For the second strategy more groups are defined by adding new levels to .groupby(). So it now includes "Sex" and "Pclass" in addition to "Title". The aggregation method is also changed, using mean() for this strategy</span></p>

```python
ageBySexClassTitle = df.groupby(['Sex','Pclass', 'Title'], as_index=False)["Age"].mean()
ageBySexClassTitle=ageBySexClassTitle.rename(columns={"Age":"Age_MeanStr2"})
ageBySexClassTitle
```

<p><span style="font-family: Helvetica; font-size: 15px;">The same procedure as before is used to include a column in the "fullAge" dataframe that fills all "Age" values by replacing nulls by the corresponding value in "ageBySexClassTitle"</span></p>
<p><span style="font-family: Helvetica; font-size: 15px;">First both tables are joined and then a new column "Age_Str2" is created by applying .combine_first() method on "Age" column with "Age_MeanStr2" as replacement column</span></p>

```python
fullAge=pd.merge(fullAge,ageBySexClassTitle,how='left',on=['Sex','Pclass', 'Title'])
fullAge["Age_Str2"]=fullAge["Age"].combine_first(fullAge["Age_MeanStr2"])
fullAge
```

<p style="text-align: left;"><span style="font-family: Helvetica; font-size: 20px;"><span style="color: rgb(0, 156, 222);"><strong>2.</strong></span></span><span style="font-family: Helvetica; font-size: 20px;">&nbsp;Compare both strategies</span></p>

Both strategies are compared in terms of Standard deviation. When calculating for the overall dataset, both values are very similar.

```python
fullAge["Age_Str1"].std()
```

```python
fullAge["Age_Str2"].std()
```

<p><span style="font-family: Helvetica; font-size: 15px;">Using groupby(), standard deviations can be computed for both strategies, grouping by "Sex" and "Pclass". It seems that there is little difference between one strategy and the other. Possibly this happens as both ways rely on segmentation by "Title", and then the additional levels introduced in second strategy don't make a relevant difference.</span></p>

```python
fullAge.groupby(["Sex", "Pclass"])[["Age_Str1", "Age_Str2"]].std()
```
